# ContLossの為の係数減衰関数の考察

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
r = 0.1
N = 5

x = np.arange(0, N)
lin_shift = -((1 - r)/(N-1)) * x + 1
exp_shift = np.exp(1/(N-1) * np.log(r) * x)
hyp_shift = 1/(1 + (1-r)/((N-1)*r)*x)
_x = np.linspace(0, N-1, 100)
print(x)
print(lin_shift)
print(exp_shift)
print(hyp_shift)
plt.plot(_x, -((1 - r)/(N-1)) * _x + 1, label='linear decay')
plt.plot(_x, np.exp(1/(N-1) * np.log(r) * _x), label='exponential decay')
plt.plot(_x, 1/(1 + (1-r)/((N-1)*r)*_x), label='invprop decay')
plt.scatter(x, lin_shift)
plt.scatter(x, exp_shift)
plt.scatter(x, hyp_shift)
plt.legend()
plt.show()
l2 = lin_shift / lin_shift.sum()
e2 = exp_shift / exp_shift.sum()
h2 = hyp_shift / hyp_shift.sum()
plt.scatter(x, l2)
plt.scatter(x, e2)
plt.scatter(x, h2)
plt.plot(x, l2, label='linear decay')
plt.plot(x, e2, label='exponential decay')
plt.plot(x, h2, label='invprop decay')
plt.legend()
plt.show()
```
