```python
import os
from os import path
import csv
import numpy as np
import matplotlib.pyplot as plt
import av
import pims
from skimage import color, measure
%matplotlib inline
```

```python
DATA_PATH = '/media/shimo/HDD_storage/DataSet/Moments_in_Time_256x256_30fps'
for p in os.listdir(DATA_PATH):
    print(p)
MOVIE_NAME = "karate.mp4"
# MOVIE_NAME = 'DRLSR_india.mp4'
```

```python
movie = pims.PyAVReaderIndexed(MOVIE_NAME)[:3]
```

```python
frame_size = 256
patch_size = 32

frame = np.array(movie)

for i in range(10):
    x = np.random.randint(frame_size - patch_size)
    y = np.random.randint(32, frame_size - patch_size - 32)

    sub_frame = frame[:, y:y+patch_size, x:x+patch_size, :]
    plt.subplot(1, 4, 1)
    plt.imshow(sub_frame[0])
    plt.subplot(1, 4, 2)
    plt.imshow(sub_frame[1])
    plt.subplot(1, 4, 3)
    plt.imshow(sub_frame[2])
    plt.subplot(1, 4, 4)
    plt.imshow(np.abs(color.rgb2gray(sub_frame[0]) - color.rgb2gray(sub_frame[2])))
    plt.show()
    psnr = measure.compare_psnr(sub_frame[0], sub_frame[2])
    var = np.var(color.rgb2gray(sub_frame[1]))
    print("PSNR", psnr)
    print("var", var)
    
```
